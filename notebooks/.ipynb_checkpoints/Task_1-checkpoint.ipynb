"""
Main goal

    1)Чтение и запись файлов с помощью pandas.
    a.Прочитать 2 файла “1. data1” и “1. data2”. (В каждом файле 2 листа).
    b.Объединить данные из файлов в один датафрейм со столбцами:
    X1, x2, y
    c.Записать исходный датафрейм в scv файл с разделителями «,»

"""

```python
# Features1
data1_sheet_1 = pd.read_excel('..\\DATA\\1. data1.xlsx',
                              index_col=None,
                              sheet_name='фичи1')
# Features2
data1_sheet_2 = pd.read_excel('..\\DATA\\1. data1.xlsx',
                              index_col=None,
                              sheet_name='фичи2')
```

```python
data1_sheet_1.head(5)
```

```python
data1_sheet_2.head(5)
```

```python
# Amswers1
data2_sheet_1 = pd.read_excel('..\\DATA\\1. data2.xlsx',
                              index_col=None,
                              sheet_name='ответы1')
# Amswers2
data2_sheet_2 = pd.read_excel('..\\DATA\\1. data2.xlsx',
                              index_col=None,
                              sheet_name='ответы2')
```

```python
# Проверка размеров датафреймов
print(f'Features1: {data1_sheet_1.shape}')
print(f'Features2: {data1_sheet_2.shape}')
print(f'Amswers1: {data2_sheet_1.shape}')
print(f'Amswers2: {data2_sheet_2.shape}')
```

```python
df1 = pd.concat([data1_sheet_1, data2_sheet_1], axis=1)
```

```python
df1.head(5)
```

```python
df2 = pd.concat([data1_sheet_2, data2_sheet_2], axis=1)
```

```python
df2.head(5)
```

```python
data = pd.DataFrame()
```

```python
df1_list = df1.values.tolist()
df2_list = df2.values.tolist()

for i in range(len(df2_list)):
    df1_list.append(df2_list[i])
full_list = df1_list
data = pd.DataFrame(full_list, columns=['x1', 'x2', 'y'])
data = data.astype({
    'x1': float,
    'x2': float,
    'y': int
})
```

```python
# Удаляем лищние данные (заботимся о памяти)
del data1_sheet_1
del data1_sheet_2
del data2_sheet_1
del data2_sheet_2
del df1
del df2
del df1_list
del df2_list
del full_list
```
